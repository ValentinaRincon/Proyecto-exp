```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["EmissionWavelength", "Intensity"]]
    return df_datos
```

```python
MSE = datos('muestra sin edta.csv')
PSE = datos('positivo sin edta.csv')
NSE = datos('negativo sin edta.csv')
```

```python
plt.scatter(MSE["EmissionWavelength"], MSE["Intensity"], marker='o', color='deeppink', label= 'Muestra')
plt.scatter(PSE["EmissionWavelength"], PSE["Intensity"], marker='o', color='chocolate', label= 'Control positivo')
plt.scatter(NSE["EmissionWavelength"], NSE["Intensity"], marker='o', color='purple', label='Control negativo')
plt.title('Espectro sin EDTA')
plt.xlabel('Longitud de Onda (nm)') 
plt.ylabel('Intensidad de Fluorescencia (Unidades Arbitrarias)')
plt.legend()
plt.grid()
plt.show()
```
