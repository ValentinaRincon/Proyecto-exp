```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos_cinetica(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "Sample", "GPolarization", "I1", "I2",]]
    return df_datos
```

# Primera toma

```python
control_1 = datos_cinetica('01_CCCP dioc cinetica.csv')

sample_1 = control_1[control_1["Sample"] == "Sample"]
sample2_1 = control_1[control_1["Sample"] == "Sample2"]
ref_1 = control_1[control_1["Sample"] == "Reference"]
```

Sample

```python
plt.plot(sample_1["Time"], sample_1["I1"], marker='o', color='royalblue', label= 'I1=500 nm')
plt.plot(sample_1["Time"], sample_1["I2"], marker='o', color='crimson', label= 'I2=540 nm')
#plt.title('Cintica 0.5 s$\muL$')
plt.xlabel('Tiempo (s)') 
plt.ylabel('Intensidad')
plt.legend(loc="center right")
plt.grid()
plt.show()

fig, axs = plt.subplots(ncols=2, figsize=(7, 2.5), constrained_layout=True)
axs[0].plot(sample_1["Time"], sample_1["I1"], marker='o', color='royalblue', label= 'I1=500 nm')
axs[0].legend(loc="lower left")
axs[0].grid()
axs[1].plot(sample_1["Time"], sample_1["I2"], marker='o', color='crimson', label= 'I2=540 nm')
axs[1].legend(loc="lower left")
axs[1].grid()
```

```python
plt.scatter(sample_1["Time"], sample_1["GPolarization"], marker='o', color='darkorange')
plt.xlabel('Tiempo (s)') 
plt.ylabel('GPolarization')
#plt.ylim(0.53,0.57)
plt.grid()
plt.show()
```
