```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["EmissionWavelength", "Intensity"]]
    return df_datos
```

```python
MCE = datos('Muestra con EDTA.csv')
PCE = datos('Positivo con EDTA.csv')
NCE = datos('Negativo con EDTA.csv')
```

```python
plt.scatter(MCE["EmissionWavelength"], MCE["Intensity"], marker='o', color='deeppink', label= 'Muestra')
plt.scatter(PCE["EmissionWavelength"], PCE["Intensity"], marker='o', color='chocolate', 
            label= 'Control positivo')
plt.scatter(NCE["EmissionWavelength"], NCE["Intensity"], marker='o', color='purple', 
            label='Control negativo')
plt.title('Espectro con EDTA')
plt.xlabel('Longitud de Onda (nm)') 
plt.ylabel('Intensidad de Fluorescencia (ua)')
plt.legend()
plt.grid()
plt.show()
```
