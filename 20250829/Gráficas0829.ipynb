```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos_cinetica(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "GPolarization", "I1", "I2",]]
    return df_datos

def datos_espectro(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["EmissionWavelength", "Intensity"]]
    return df_datos
```

```python
CINE1 = datos_cinetica('intento01_CCCP_cinetica.csv') #intento sin corriente jaja
CINE2 = datos_cinetica('intento02_CCCP_cinetica.csv') #intento correcto
MSC = datos_espectro('Espectro_sin_cccp.csv')
MCC = datos_espectro('Espectro_con_cccp.csv')
```

## Cinética

```python
plt.scatter(CINE1["Time"], CINE1["I1"], marker='o', color='lightcoral', label= 'I1=500 nm')
plt.scatter(CINE1["Time"], CINE1["I2"], marker='o', color='chocolate', label= 'I2=540 nm')
#plt.title('Cinética 0.5 s$\muL$')
plt.xlabel('Tiempo (s)') 
plt.ylabel('Intensidad')
plt.legend()
plt.grid()
plt.show()
print("puntos escogidos para el caso de DiOC 2 muL: " + str(500) + "nm y " + str(540) + "nm")
```

```python
plt.scatter(CINE1["Time"], CINE1["GPolarization"], marker='o', color='darkorange')
plt.xlabel('Tiempo (s)') 
plt.ylabel('GPolarization')
#plt.legend()
plt.grid()
plt.show()
```
