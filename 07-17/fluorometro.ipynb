Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

Datos

```python
def limpiar_valor(valor):
    valor = valor.strip()
    partes = valor.split('.')
    
    if len(partes) <= 2:
        # Tiene uno o ningún punto → está bien
        return valor
    else:
        # Más de un punto → unimos todo menos el último con '' y dejamos el último como decimal
        entero = ''.join(partes[:-1])
        decimal = partes[-1]
        return f"{entero}.{decimal}"

def datos(csv_file_path):
    tabla = []
    with open(csv_file_path, newline='', encoding='utf-8') as archivo:
        lector = csv.reader(archivo, delimiter=';')
        for fila in lector:
            fila_limpia = []
            for celda in fila:
                valor = limpiar_valor(celda)
                try:
                    fila_limpia.append(float(valor))
                except ValueError:
                    fila_limpia.append(valor.strip())  # No es número, mantenemos como texto
            tabla.append(fila_limpia)
    return tabla
```

Muestra a 4°C sin diluir

```python
M4C = datos('Muestra_4C sin EDTA.csv')
P4C = datos('Positivo_4C sin EDTA.csv')
```

Muestra a 37°C sin diluir

```python
M37C = datos('Muestra_37C sin EDTA.csv')
P37C = datos('Positiva_37C sin EDTA.csv')
```

Muestra a 4°C diluida

```python
M4C_d = datos('Muestra_4C_2ml PBS sin EDTA.csv')
P4C_d = datos('Positivo_4C_2ml PBS sin EDTA.csv')
```

Muestra a 37°C diluida

```python
M37C_d = datos('Muestra_37C_2ml PBS sin EDTA.csv')
P37C_d = datos('Positivo_37C_2ml PBS sin EDTA.csv')
```

```python
def longitud_onda(dato):
    return [row[1] for row in dato]
```

```python
def intensidad_fluorescencia(dato):
    return [float(row[2]) for row in dato]
```

```python
# Muestra a 4°C sin diluir

plt.plot(longitud_onda(M4C[:130]), intensidad_fluorescencia(M4C[:130]), marker='o', linestyle='-', color='b', label = 'Muestra')
plt.plot(longitud_onda(P4C[:130]), intensidad_fluorescencia(P4C[:130]), marker='o', color='#9BF3F0', label='Control positivo')
plt.title('Intensidad de Fluorescencia vs Longitud de Onda (4°C sin EDTA)')
plt.xlabel('Longitud de Onda (nm)') 
plt.ylabel('Intensidad de Fluorescencia (Unidades Arbitrarias)')
#plt.grid()
plt.legend()
plt.show()
```
