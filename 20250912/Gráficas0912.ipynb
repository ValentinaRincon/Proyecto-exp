```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos_negativos(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "Sample", "GPolarization", "I1", "I2",]]
    return df_datos

def datos_cinetica(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "GPolarization", "I1", "I2",]]
    return df_datos

def datos_espectro(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["EmissionWavelength", "Intensity"]]
    return df_datos
```

# Controles Negativos

Con 2 mallas y rendijas de 0.5 para la de atrás  y 1 en todas las demás

```python
control_2 = datos_negativos('05 con y sin CCCP cinetica (con bacterias).csv')

sample = control_2[control_2["Sample"] == "Sample"]
sample2 = control_2[control_2["Sample"] == "Sample"]
```

```python
sample = control_2[control_2["Sample"] == "Sample"]
sample2 = control_2[control_2["Sample"] == "Sample"]
```

```python
plt.scatter(sample["Time"], sample["I1"], marker='o', color='royalblue', label= 'I1=500 nm')
plt.scatter(sample["Time"], sample["I2"], marker='o', color='crimson', label= 'I2=540 nm')
#plt.title('Cinética 0.5 s$\muL$')
plt.xlabel('Tiempo (s)') 
plt.ylabel('Intensidad')
plt.legend(loc="center right")
plt.grid()
plt.show()

fig, axs = plt.subplots(ncols=2, figsize=(7, 2.5), constrained_layout=True)
axs[0].plot(sample["Time"], sample["I1"], marker='o', color='royalblue', label= 'I1=500 nm')
axs[0].legend(loc="lower right")
axs[0].grid()
axs[1].plot(sample["Time"], sample["I2"], marker='o', color='crimson', label= 'I2=540 nm')
axs[1].legend(loc="lower right")
axs[1].grid()
```

```python
plt.scatter(sample["Time"], sample["GPolarization"], marker='o', color='darkorange')
plt.xlabel('Tiempo (s)') 
plt.ylabel('GPolarization')
#plt.ylim(0.53,0.57)
plt.grid()
plt.show()
```
