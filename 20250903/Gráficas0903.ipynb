```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos_cinetica(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "GPolarization", "I1", "I2",]]
    return df_datos

def datos_espectro(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["EmissionWavelength", "Intensity"]]
    return df_datos
```

```python
CINE1 = datos_cinetica('cinetica_550.csv')
CINE2 = datos_cinetica('cinetica02_550.csv')
CINE3 = datos_cinetica('Cinetica_540.csv')
NEG = datos_cinetica('Control negativo_cinetica_540.csv')
MSC = datos_espectro('Espectro_sin_CCCP.csv')
MCC = datos_espectro('Espectro_con_CCCP .csv')
```

Primera prueba 550

```python
plt.scatter(CINE1["Time"], CINE1["I1"], marker='o', color='royalblue', label= 'I1=500 nm')
plt.scatter(CINE1["Time"], CINE1["I2"], marker='o', color='crimson', label= 'I2=550 nm')
#plt.title('Cin√©tica 0.5 s$\muL$')
plt.xlabel('Tiempo (s)') 
plt.ylabel('Intensidad')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter(CINE1["Time"], CINE1["GPolarization"], marker='o', color='darkorange')
plt.xlabel('Tiempo (s)') 
plt.ylabel('GPolarization')
#plt.legend()
plt.grid()
plt.show()
```
