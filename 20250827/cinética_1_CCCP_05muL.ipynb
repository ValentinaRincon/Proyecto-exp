```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def datos(archivo_csv):
    df = pd.read_csv(archivo_csv, sep=';')
    df_datos = df[["Time", "GPolarization", "I1", "I2"]]
    return df_datos
```

```python
CIN = datos('CCCP_2muL DiOC cin√©tica_1.csv')
```

```python
plt.scatter(CIN["Time"], CIN["I1"], marker='o', color='darkorange', label= 'I1')
plt.scatter(CIN["Time"], CIN["I2"], marker='o', color='brown', label= 'I2')
#plt.title('Cin√©tica 0.5 s$\muL$')
plt.xlabel('Tiempo (s)') 
plt.ylabel('Intensidad')
plt.legend()
plt.grid()
plt.show()
print("puntos escogidos para el caso de DiOC 2 muL: " + str(500) + "nm y " + str(520) + "nm")
```

```python
plt.scatter(CIN["Time"], CIN["GPolarization"], marker='o', color='darkorange')
plt.xlabel('Tiempo (s)') 
plt.ylabel('GPolarization')
#plt.legend()
plt.ylim(0.0,0.4)
plt.grid()
plt.show()
```
